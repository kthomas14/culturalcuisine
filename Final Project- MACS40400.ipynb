### Introduction and Peircian Semiotics

Why are some cultural cuisines more accommodating to dietary restrictions than others? My research question aims to map the accessibility of plant-based/alternative diets within different cultures in the US. Many races and ethnicities have brought rich culture and cuisine to the US that stays true to the country from which it originated. With the growing popularity of alternative diets like veganism and vegetarianism, some cultures are naturally adept at providing accommodations for these restrictions, while other cultures do not have as accessible or common accommodations for those with alternative diets. The goal of this research question is to assess the availability of dishes that meet alternative diet requirements across different cultures, and then to examine why there may be more accommodation from some cultures rather than others. Many cultural factors may contribute to the explanations as to why some cuisines provide more alternative options than others including, religious restrictions, food or crop availability, and overall cultural preferences. The data collected for the purposes of this research is originated from US cities, so American culture is also incorporated into these patterns. Due to the rising popularity of plant-based diets in the United States, restaurants may be motivated to provide plant-based items for patrons while still adhering to the cultural patterns of their respective cuisine. Whether these accommodations come naturally or are provided in order to appeal to a new audience, the extent to which these options are available will be assessed through the number of plant-based, vegan, or vegetarian items listed or described on the menu. For ease of understanding, the term plant-based will be from this point forward in reference to plant-based, vegan, and vegetarian diets as a collective.

Each culture may replicate the legisign of what constitutes a plant-based diet differently. These distinctions are what separates each culture from one another. While these replications vary between culture, the key properties which must be replicated and interpreted is consistent among all cultures. The overall replication is being performed on a legisign. In Peircean semiotics, this is a sign which encompasses a cultural norm or law within its representation. The legisign being replicated within each culture is the qualities that make a food plant based. Additionally, these qualities are the object of the sign. The qualities and norms with which an item must follow in order to be considered plant-based is the object within this legisign. These qualities are being replicated with each sign created in a culture. The sign-vehicle is the actual replication of the object. In the analysis of menu items, which is what occurs in this study, the sign-vehicle is the description of menu items that signify the dish is plant-based. This sign vehicle is referring to the object symbolically. A menu description that follows the criterion of a plant-based diet is represented using the conventions of the object in question. This follows the semiotic definition of symbolism. The interpretant of the sign-vehicle is the presented food item. Since the interpretant is viewing the sign-vehicle as a replication of the object itself, the sign-vehicle is rhematically representing its qualities to the interpretant.

```python
# Combining each Menu Item with its Description
def combine_items_and_desc(menus):
    menus['Items and Descriptions'] = ''
    for men_index, menu in menus.iterrows():
        items_and_desc = []
        for index, items in enumerate(menu['Menu Items']):
            item_and_desc_text = str(menu['Menu Items'][index]) + str(menu['Item Descriptions'][index])
            items_and_desc.append(item_and_desc_text)
        menus.at[men_index, 'Items and Descriptions'] = items_and_desc
```

**Creating index**

* An index was created in order to find replications of the legisign within restuaurant menus.
* The data is then collected and the number of Plant Based Alternatives per restaurant and cuisine are used in order to start analysis of how accessible alternative diets are in differing cultures. 

```python
# creating word index to search for plant based options
diet_regex = 'veggie|vegan|vegetarian|plant based|plant-based|impossible burger|beyond burger|beyond meat|setian|tempeh|chickn|chikn|mylk|almond milk|oat milk'

# tally for plant based alternatives within a menu
def plant_options(menus, regex):
    menus['Plant Based Alternatives'] = ''
    for index, row in menus.iterrows():
        word_freq = 0
        for item in row['Items and Descriptions']:
            if re.search(regex, item, re.I):
                word_freq += 1
        row['Plant Based Alternatives'] = word_freq
    menus['Plant Based Alternatives'] = menus['Plant Based Alternatives'].astype(int)
```

```python
plant_options(menus_ny, diet_regex)
menus_ny.head()
```

```python
plant_options(menus_orl, diet_regex)
menus_orl.head()
```

```python
# Prepping to plot histogram for word freq
# grouping by cuisine

def cuisine_grouping(menus):
    cuisine = pd.DataFrame({'Cuisine': cuisine_lst, 'num_options': 0, 'avg_options': 0})
    cuisine_avg = menus.groupby('Cuisine').mean('Plant Based Alternatives')
    cuisine_median = menus.groupby('Cuisine').median('Plant Based Alternatives')
    for cus_index, cus_row in cuisine.iterrows():
        for men_index, men_row in menus.iterrows():
            if men_row['Cuisine'] == cus_row['Cuisine']:
                cuisine.loc[cus_index,'num_options'] += men_row['Plant Based Alternatives']

    for cus_index, cus_row in cuisine.iterrows():
        for cus_avg_index, cus_avg_row in cuisine_avg.iterrows():
            if cus_row['Cuisine'] == cus_avg_index:
                cuisine.loc[cus_index, 'avg_options'] = cus_avg_row[0]
            else:
                continue
        for cus_med_index, cus_med_row in cuisine_median.iterrows():
            if cus_row['Cuisine'] == cus_med_index:
                cuisine.loc[cus_index, 'median_options'] = cus_med_row[0]
            else:
                continue
    return cuisine
```

**Data Summary for New York**

* The following cells display summary data describing how many plant based options are offered in the New York area per cuisine. 
* The folium package was also used in order to create maps displaying the distribution of the restaurants included in the data base. The interactive map allows for users to select a datapoint in order to view the restaurant name, the type of cuisine offered, and how many plant based alternatives that specific restaurant offers. The following heat map displays the density of plant based alternatives in New York. The red areas indicate that restaurants in that area offer a large amount of alternatives. These maps can be used in tandem to determine which areas offer the most alternaives, and what types of cuisines are offered in that area. 

```python
# Plots for New York
cuisine_ny = cuisine_grouping(menus_ny)

fig, axes = plt.subplots(nrows=1, ncols=2)
cuisine_ny.plot(x = 'Cuisine', y = 'num_options', 
                ax=axes[0], kind='bar', figsize = (20, 6), width = 0.90)

cuisine_ny.plot(x = 'Cuisine', y = 'avg_options', ax=axes[1], 
                color = 'orange', kind='bar', figsize = (20, 6), width = 0.90)
cuisine_ny.sort_values('avg_options', ascending = False)
```
